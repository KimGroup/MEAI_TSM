```python
import xlrd
#import xlwt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_excel("dsq_d_fcc.xls")
```

```python
tsm=df['TSM'].dropna().to_numpy()
tri=df['Trivial'].dropna().to_numpy()
```

```python
tsm.shape,tri.shape
```

```python
np.min(tsm),np.min(tri),np.max(tsm),np.max(tri)
```

```python
np.mean(tsm)
```

```python
# libraries & dataset
import seaborn as sns
import numpy as np
#sns.set(style="darkgrid")

plt.figure(figsize=(10,2))


sns.histplot(tsm, bins=50,binrange=(0.5,1.15), color="teal", label="TSM", kde=False,element='step')
sns.histplot(tri, bins=50,binrange=(0.5,1.15), color="orange", label="Trivial", kde=False,element='step')

plt.plot([0.677,0.677], [0.5,110],lw=2.5,linestyle='--',color='black')
plt.ylim([0.5,120])

plt.yscale('log')

plt.rcParams['font.family']='Times New Roman'

plt.xticks(np.arange(0.5,1.15,0.2),size=30)
plt.yticks(ticks=[1,10,100],labels=['1','10','100'],size=30)

plt.xlabel('$\it{d_{sq}/fcc}$',size=35)
plt.ylabel(r'Counts',size=35)


plt.minorticks_off()

plt.legend(fontsize=30,frameon=False,labelspacing=0.3,handlelength=1.5,handletextpad=0.2,borderaxespad=0.01) 
plt.savefig(r'dsq_fcc_classification.svg',bbox_inches='tight',transparent='True')
plt.show()
```
