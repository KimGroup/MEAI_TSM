```python
def sign(number):
    if number>=0:
        return 1.
    else:
        return -1.

def rescale(matrix):

    for ri in range(len(matrix)):
        for rj in range(len(matrix[ri])):
            if abs(matrix[ri][rj])>0.85:
                matrix[ri][rj]=sign(matrix[ri][rj])*0.475+sign(matrix[ri][rj])*(abs(matrix[ri][rj])-0.85)*5.25/1.5
            else:
                matrix[ri][rj]=matrix[ri][rj]/2.
    return matrix
```
