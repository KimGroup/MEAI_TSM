# Leslie's comments (Mar 13)




InTe: The rocksalt structure is not the preferred structure, it is usually tetragonal. The cubic form exists at high pressure. It is a semimetal because of electron count. But it is an unstable phase. We could put it with a weaker color or something to indicate that this is only a metastable phase

BiSe: Same thing, unstable phase, I’m not even convinced it actually exists. Based on electron count it is a semimetal, but I would also somehow indicate in the plot that this might not be stable.

HgTe: I would label is as TI because it was the first TI we found

InAs: definitely trivial, very well known semiconductor, is also a known example that DFT gets wrong

SnAs: This one is indeed odd, it is stable, and a semimetal based on electron count. It is also a superconductor. https://pubs.acs.org/doi/abs/10.1021/cm503992d?casa_token=fKBbRmLmrckAAAAA:swV9O-2MOCqUURbiVH5uI48GqQWZCrzlgbEWAIWF3Q6aiBDn65kCkWGCCSHdStL-81t-X6LOTnT8x6Y I think we need to keep it as is.

InSb: trivial. Also a known semiconductor with no band inversion, just like InAs

BiTe: same as BiSe

SbSn: High pressure phase in rock salt. Prefers other structure. Same as BiSe

SnP: same as above

AlO: extremely unstable phase, was seen in a crazy experiment once (10.1021/ja01638a076). Would be SM based on electron count but definitely not stable

ZnTe: the existence of this is very questionable. It usually crystallizes in the wurtzide, not rock salt structure. There is one report from 1926 which says its rocksalt but it might have just been wrong. There. Is another French one form 1983 to which I don’t have access. But it can’t be the thermodynamically stable phase.

TlC: I don’t know where this comes form, it does not exist in the ICSD if I search for it. Maybe just delete?

# Precision recall

```python
len(Prob)
```

```python
y_label=label[(label=='TI')|(label=='TV')|(label=='SM')|(label=='trivial/SM')]
y_label=[1 if i=='TI' else 0 for i in y_label]



y_prob=Prob[(label=='TI')|(label=='TV')|(label=='SM')]

y_pred=[1 if i>0.5 else 0 for i in y_prob]
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


conf_matrix = confusion_matrix(y_label, y_pred)


# Normalize the confusion matrix by row (i.e., by the number of true classes)
conf_matrix_normalized = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]

# Display the normalized confusion matrix using seaborn for better visualization
plt.figure(figsize=(5, 5))

annot_labels=[str(i)+'%' for i in np.round(conf_matrix_normalized.flatten(),1)]

sns.heatmap(conf_matrix_normalized, annot=True,annot_kws={'size': 25},cbar=False, fmt='.1%', cmap='Blues', 
            xticklabels=['Trivial/SM', 'TI'], yticklabels=['Trivial/SM', 'TI'])

plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

#ax=plt.gca()
#ax.xaxis.set_label_position('top') 
#ax.xaxis.set_ticks_position('top') 


plt.xlabel('ME-AI predictions',size=25,labelpad=20)
plt.ylabel('True labels',size=25,labelpad=20)
plt.xticks(size=23)
plt.yticks(size=23)

#plt.title('Generalized ME-AI\n',size=30)

plt.gca().spines['left'].set_visible(True)
plt.gca().spines['right'].set_visible(True)
plt.gca().spines['top'].set_visible(True)
plt.gca().spines['bottom'].set_visible(True)


plt.savefig(r'Rocksalt_confusion_normalized.svg',bbox_inches='tight',transparent='True')
plt.show()

```
